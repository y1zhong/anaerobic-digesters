### Calculate cross-correlation and time lag between microbe and chemical measurements

```r
library(stringr)
library(tseries)
library(dplyr)
```

```r
Y.m <- read.csv('trimmed-150nts/correlation-analysis/alr_otu_taxon_table.txt', sep='\t', row.names=1, check.names = F)
mf <- read.csv('trimmed-150nts/correlation-analysis/alr-mapping-table.txt', sep='\t', row.names=1)
head(Y.m)
head(mf)
```

extract taxanomy assignment infomation

```r
taxa <- data.frame(rownames(Y.m), Y.m[, 1])
n_taxa <- nrow(taxa)
colnames(taxa) <- c('sequence', 'assignment')

taxa$names <- paste(rep('Bac', n_taxa), rownames(taxa), sep = '')
taxa$kingdom <- sapply(1:nrow(taxa), function(i){
  if(as.character(taxa$assignment[i]) == 'Unassigned') return('Unassigned')
  extr <- str_extract(as.character(taxa$assignment[i]), "(?<=k__).*?(?=;)")
  if(!is.na(extr)) {
    extr
  }else{
    strsplit(as.character(taxa$assignment[i]), "k__")[[1]][2]
  }
})

taxa$phylum <- sapply(1:nrow(taxa), function(i){
  if(nchar(as.character(taxa$assignment[i])) <= 11) return('no_phylum_info')
  extr <- str_extract(as.character(taxa$assignment[i]), "(?<=p__).*?(?=;)")
  if(!is.na(extr)) {
    extr
  }else{
    strsplit(as.character(taxa$assignment[i]), "p__")[[1]][2]
  }
})
head(taxa)
```

```r
bac <- data.frame(t(Y.m[2:ncol(Y.m)]))
bac <- data.frame(apply(bac, 2, function(x) as.numeric(as.character(x))), row.names=rownames(bac))
colnames(bac) <- taxa$names
bac$id <- sapply(1:dim(bac)[1], function(i){ 
  strsplit(rownames(bac)[i], "-")[[1]][1]
})
bac$Day <- sapply(1:dim(bac)[1], function(i){ 
  as.numeric(strsplit(rownames(bac)[i], "-")[[1]][2])
})
bac <- bac[with(bac, order(id, Day)), ]
head(bac)
```

```r
chems <- mf[, c("ph","cod","phosphorousr","phosphoroust" ,"ammonian", 
                "alkalinity","nitraten", "nitrogent" ,"volatile_fatty_acids" ,"tss_vss")]
chem_names <- c("pH","COD","PhosphorousR","PhosphorousT" ,"AmmoniaN", 
                "Alkalinity","NitrateN", "NitrogenT" ,"Volatile.Fatty.Acids" ,"TSS.VSS")
n_chem <- length(chem_names)
colnames(chems) <- chem_names
chems <- data.frame(apply(chems, 2, function(x) as.numeric(as.character(x))), row.names=rownames(chems))
chems$id <- sapply(1:dim(chems)[1], function(i){
  strsplit(rownames(chems)[i], "-")[[1]][1]
})
chems$Day <- sapply(1:dim(chems)[1], function(i){ 
  as.numeric(strsplit(rownames(chems)[i], "-")[[1]][2])
})
chems <- chems[with(chems, order(id, Day)), ]
```

separate data by digseter and order by date

```r
di_ids <- names(table(chems$id))
chem_all <- lapply(1:4, function(i)
  chem_temp <- chems[which(chems$id == di_ids[i]), chem_names])
bac_all <- lapply(1:4, function(i)
  bac_temp <- bac[which(bac$id == di_ids[i]), taxa$names])
```

```r
head(chem_all[[1]])
head(bac_all[[1]])
```

```r
save(taxa, chem_all, bac_all, file='trimmed-150nts/correlation-analysis/processed_data.RData')
```

get cross-correlation with max time lag 1

```r
corr <- list()
lagg <- list()
for(i in 1:4){
  chem_temp <- chem_all[[i]]
  bac_temp <- bac_all[[i]]
  na_days <- rownames(chem_temp)[which(is.na(chem_temp[,1]))]
  
  #delete na values
  chem_omit <- chem_temp[!rownames(chem_temp) %in% na_days, ]
  bac_omit <- bac_temp[!rownames(bac_temp) %in% na_days, ]
  
  corr_temp <- data.frame(matrix(0,nrow = n_taxa, ncol = n_chem))
  rownames(corr_temp) <- colnames(bac_omit)
  colnames(corr_temp) <- colnames(chem_omit)
  lagg_temp <- data.frame(matrix(0,nrow = n_taxa, ncol = n_chem))
  rownames(lagg_temp) <- colnames(bac_omit)
  colnames(lagg_temp) <- colnames(chem_omit)
  for(j in c(1:n_chem)){
    co_la_onechem <- sapply(1:n_taxa, function(k){
      ccf_temp <- ccf(chem_omit[,j], bac_omit[,k], lag.max = 1, plot = F)
      ccr = ccf_temp$acf[,,1]
      lag = ccf_temp$lag[,,1]
      corr_temp <- ccr[which.max(abs(ccr))]
      lagg_temp <- lag[which.max(abs(ccr))]
      c(corr_temp, lagg_temp)
    })
    corr_temp[, j] <- co_la_onechem[1, ]
    lagg_temp[, j] <- co_la_onechem[2, ]
  }
  corr[[i]] <- corr_temp
  lagg[[i]] <- lagg_temp
}
```

```r
head(corr[[1]])
head(lagg[[1]])
```

```r
save(corr,lagg, file = 'trimmed-150nts/correlation-analysis/cor_lag.RData')
```
