### Additive log-ratio transformation on otu file

```r
#change path
otu = read.csv('trimmed-150nts/correlation-analysis/feature-table.txt', sep='\t', skip=1)
taxonomy = read.csv('trimmed-150nts/correlation-analysis/taxonomy/taxonomy.tsv', sep='\t')
```

```r
head(otu)
head(taxonomy)
```

```r
n_otu <- ncol(otu) - 1
colnames(otu)[1] = 'Feature.ID'
otu_taxon = merge(otu, taxonomy, by="Feature.ID")
rownames(otu_taxon) = otu_taxon[, 1]
otu_taxon = otu_taxon[, -1]
dim(otu_taxon)
head(otu_taxon)
```

```r
colnames(otu_taxon)[1:n_otu] <- sub(".*?\\.", "", colnames(otu_taxon)[1:n_otu]) 
colnames(otu_taxon)[1:n_otu]
```

```r
otu_d1 = otu_taxon[,grepl("d1", colnames(otu_taxon))]
otu_d2 = otu_taxon[,grepl("d2", colnames(otu_taxon))]
otu_d3 = otu_taxon[,grepl("d3", colnames(otu_taxon))]
otu_d4 = otu_taxon[,grepl("d4", colnames(otu_taxon))]
head(otu_d1)
```

Only keep OTU exists in at least 3 days for each water digester

```r
otu_d1 = otu_d1[rowSums(otu_d1 > 0) >= 3, ]
otu_d2 = otu_d2[rowSums(otu_d2 > 0) >= 3, ]
otu_d3 = otu_d3[rowSums(otu_d3 > 0) >= 3, ]
otu_d4 = otu_d4[rowSums(otu_d4 > 0) >= 3, ]

dim(otu_d1)
dim(otu_d2)
dim(otu_d3)
dim(otu_d4)
```

keep common taxa in four digesters

```r
otu_selected = Reduce(intersect, list(rownames(otu_d1), rownames(otu_d2), rownames(otu_d3), rownames(otu_d4)))
length(otu_selected)

otu_taxon_sub = otu_taxon[rownames(otu_taxon) %in% otu_selected, ]
dim(otu_taxon_sub)
head(otu_taxon_sub)
```

select spike-in taxa and use the highset one as reference in alr transformation

```r
spikes = otu_taxon_sub[grepl('g__Acinetobacter', otu_taxon_sub$Taxon), ] 
dim(spikes)
head(spikes, 3) # values vary differently: choose the highest one
rowSums(spikes[, 1:n_otu])
```

```r
spikes_highest = spikes[which.max(rowSums(spikes[, 1:n_otu])),]
head(spikes_highest)
sum(spikes_highest[, 1:n_otu])
rownames(spikes_highest)
```

remove all spikes

```r
alr_otu_taxon_sub = otu_taxon_sub[!(rownames(otu_taxon_sub) %in% rownames(spikes)), ] 
dim(alr_otu_taxon_sub)
head(alr_otu_taxon_sub)
```

additive log-ratio transformation with 0.1 pseudocount

```r
for (i in 1:dim(alr_otu_taxon_sub)[1]){
  alr_otu_taxon_sub[i, 1:n_otu][which(alr_otu_taxon_sub[i, 1:n_otu] == 0)] = 0.1 
  alr_otu_taxon_sub[i, 1:n_otu] = log(alr_otu_taxon_sub[i, 1:n_otu]/spikes_highest[, 1:n_otu])
}
head(alr_otu_taxon_sub)
```

```r
colnames(alr_otu_taxon_sub) = toupper(gsub('.', '-', colnames(alr_otu_taxon_sub), fixed=T)) 
head(alr_otu_taxon_sub)
```

### Mapping file

```r
mf = read.csv('mapping-file.tsv', sep='\t', row.names=1)
rownames(mf) <- sub(".*?\\.", "", rownames(mf)) 
mf = mf[!rownames(mf) %in% c('d0', 'd0.spike'), ]
rownames(mf) = toupper(gsub('.', '-', rownames(mf), fixed=T)) 
head(mf)
```

```r
setdiff(colnames(alr_otu_taxon_sub), rownames(mf))
```

```r
sample_ids = intersect(colnames(alr_otu_taxon_sub), rownames(mf))
sample_ids = sort(sample_ids) 
length(sample_ids)
```

```r
alr_otu_taxon_final = alr_otu_taxon_sub[, c('TAXON', sample_ids)]
mf = mf[sample_ids, ]
```

```r
dim(alr_otu_taxon_final); 
dim(mf)
head(alr_otu_taxon_final)
head(mf)
```

```r
write.table(alr_otu_taxon_final, file='trimmed-150nts/correlation-analysis/alr_otu_taxon_table.txt', sep='\t')
write.table(mf, file='trimmed-150nts/correlation-analysis/alr-mapping-table.txt', sep='\t', col.names=NA)
```
